```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# i would load  the dataset
data = pd.read_excel('/content/Dataset5.xlsx')

#then i would  extract relevant columns for average monthly searches
keyword_searches = data[['Keyword', 'Avg. monthly searches']].dropna()




```

```python
# Extracting relevant columns for regional interest scores
regional_interest = data[['Region', 'Interest Score.2']].dropna()

# Creating a vertical bar graph for Regional Interest Scores
plt.figure(figsize=(10, 6))
sns.barplot(x='Interest Score.2', y='Region', data=regional_interest.sort_values('Interest Score.2', ascending=False))
plt.title('Regional Interest Scores for Leffe')
plt.xlabel('Interest Score')
plt.ylabel('Region')
plt.show()
```

```python
keyword_searches = data[['Keyword', 'Avg. monthly searches']].dropna()
regional_interest = data[['Region', 'Interest Score.2']].dropna()

# Sorting the data by 10 keywords by average monthly searches
top_keywords = keyword_searches.sort_values(by='Avg. monthly searches', ascending=False).head(10)


# Setting  the seaborn style
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(x='Avg. monthly searches', y='Keyword', data=top_keywords, palette="Blues_d")
plt.title('Top 10 Keywords by Average Monthly Searches')
plt.xlabel('Average Monthly Searches')
plt.ylabel('Keywords')
plt.show()
```
